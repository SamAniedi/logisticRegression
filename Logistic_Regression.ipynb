# **DATA CLEANING**

```python
df.isnull().sum()
```

```python
df2.isnull().sum()
```

```python
import seaborn as sns
sns.histplot(data=df, x='Sex')
```

```python
sns.histplot(data=df2, x='Sex')
```

```python
df.drop('Cabin', axis=1, inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
df.isnull().sum()
```

```python
df2.drop('Cabin', axis=1, inplace=True)
df2['Age'].fillna(df2['Age'].median(), inplace=True)
df2['Fare'].fillna(df2['Fare'].median(), inplace=True)
```

```python
df2.isnull().sum()
```

```python
encoded_vars = pd.get_dummies(data=df, columns=['Sex', 'Name', 'Ticket', 'Embarked'], drop_first=True)
encoded_vars = encoded_vars.drop(['PassengerId'], axis=True)
encoded_vars.head()
```

```python
numeric_vars = encoded_vars[['Survived','Age','SibSp', 'Parch']]
numeric_vars
```

```python
encoded_vars2 = pd.get_dummies(data=df2, columns=['Sex','Name', 'Ticket', 'Embarked'], drop_first=True)
encoded_vars2 = encoded_vars2.drop('PassengerId', axis=True)
encoded_vars2.head()
```

```python
numeric_vars2 = encoded_vars2[['Age','SibSp', 'Parch']]
numeric_vars2
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

# **DATA PREPROCESSING**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(numeric_vars)
scaled_df = pd.DataFrame(scaled_data, columns=numeric_vars.columns)
scaled_df.head()
```

```python
temp_df = encoded_vars.drop(numeric_vars.columns, axis=True)
prep_df = pd.concat([scaled_df, temp_df], axis=1)
prep_df
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler2 = MinMaxScaler()
scaled_data2 = scaler2.fit_transform(numeric_vars2)
scaled_df2 = pd.DataFrame(scaled_data2, columns=numeric_vars2.columns)
scaled_df2.head()
```

```python
temp_df2 = encoded_vars2.drop(numeric_vars2.columns, axis=True)
prep_df2 = pd.concat([scaled_df2, temp_df2], axis=1)
prep_df2.head()
```

```python
X = prep_df[['Age','SibSp', 'Parch']]
X
```

```python
X2 = prep_df2[['Age','SibSp', 'Parch']]
X2
```

```python
y = prep_df['Sex_male']
y
```

```python
y2 = prep_df2['Sex_male']
y2
```

# **MODEL CREATION**

```python
r_state = 30
t_size = 0.30
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=t_size, random_state=r_state)

print('The shape of X_train: ', X_train.shape)
print('The shape of X_test: ', X_test.shape)
print('The shape of y_train: ', y_train.shape)
print('The shape of y_test: ', y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegressionCV
model2 = LogisticRegressionCV(cv=5, random_state=0)
model2.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model2.coef_
```

```python
model.intercept_
```

```python
model2.intercept_
```

# **MODEL EVALUATION**

```python
pred = X2.iloc[:y_test.size,:]
```

```python
y_pred = model.predict(pred)
```

```python
y_pred2 = model2.predict(pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred, target_names=['test','pred']))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred2, target_names=['test','pred']))
```

```python
model.score(X2, y2)
```

```python
model2.score(X2, y2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
